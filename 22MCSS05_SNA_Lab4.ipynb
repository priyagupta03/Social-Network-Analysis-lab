1. Consider a scenario where 50 people have been invited to a party and don’t know 
each other. They can randomly interact with each other at the party. Use the network 
package  to  generate  a  random  network  that  models  the  above  behaviour.  Each 
person  in  the  network  is  considered  a  node  and  interaction  as  a  link.  Further, 
generate the random networks for the different values of probability p and degree 
k. Also, plot the degree distribution curves and analyze the results. [ Don’t use the 
Erdos_Renyi model for generating the network] 


```python
import networkx as nx
import random
import matplotlib.pyplot as plt
```

```python
n = 50
p = 0.25

nodes = [i for i in range(1, n+1)]
edges = [(i, j) for i in range(1, n+1) for j in range(i+1, n+1) if random.random()<= 0.25]
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(nodes)
G.add_edges_from(edges)
nx.draw(G, with_labels = True)
plt.show()
```

```python
import matplotlib.pyplot as plt

def plot_degree_dist(G):
    degrees = [G.degree(n) for n in G.nodes()]
    plt.hist(degrees)
    plt.show()
    print(degrees)
    plt.plot(degrees)
    plt.title("Degree Distribution")
    plt.xlabel("Node")
    plt.ylabel("Degree")
    plt.show()
plot_degree_dist(G)
```

```python
for i in range(0,3):
  n = int(input('Enter number of nodes:'))
  p = float(input('Enter probability'))
  nodes = [i for i in range(1, n+1)]
  edges = [(i, j) for i in range(1, n+1) for j in range(i+1, n+1) if random.random()<= p]
  G1 = nx.Graph()
  G1.add_nodes_from(nodes)
  G1.add_edges_from(edges)
  nx.draw(G1, with_labels = True)
  plt.show()
  plot_degree_dist(G1)
```

3. Create a regular lattice with 10 nodes and 4 edges, where each node is connected to 
4 of its nearest neighbours. Further, rewire the network using Watts-Strogatz model 
and  generate  a  small  world  model.  Generate  different  small-world  networks  for 
different values of rewiring probability pr and plot degree distribution curves. 

```python
G3 = nx.random_regular_graph(4, 10)
nx.draw(G3, with_labels = True)
plot_degree_dist(G3)

probability = [random.random() for _ in range(5)]
for p in probability:
  G4 = nx.watts_strogatz_graph(10,4,p, seed= None)
  nx.draw(G4, with_labels = True)
  plot_degree_dist(G4)
```
