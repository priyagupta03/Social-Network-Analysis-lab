3. Generate a scale-free network for different degrees of the incoming node and analyse the degree distribution curves.

```python
import matplotlib.pyplot as plt

def plot_degree_dist(G):
    degrees = [G.degree(n) for n in G.nodes()]
    plt.plot(degrees)
    plt.title("Degree Distribution")
    plt.xlabel("Node")
    plt.ylabel("Degree")
    plt.show()
```

```python
for degree in range(1,5):
  G3 = nx.barabasi_albert_graph(5, degree)
  nx.draw(G3)
  plot_degree_dist(G3)
```
