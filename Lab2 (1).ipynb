Q4.Use networkx package and generate rode network of the India’s cities. To generate the network, first create a list that stores the name of different cities and then create a list of weighted edges. After that edge is added between randomly selected city c1 and c2. Please note c1 and c2 should not be same with no existing edge.

```python
G4 = nx.Graph()
```

```python
city_list = ['Kolkata', 'Gaziabad', 'Surat', 'Mumbai', 'Allahbad', 'Jaipur', 'Ahmedabad', 'Udhampur', 'Channai', 'Hyderabad','Banglore', 'Pune', 'Lucknow', 'Mumbai', 'Kolkata']
```

```python
G4.add_nodes_from(city_list)
G4.nodes()
```

```python
edge_list = [28,44,32,66,74,11,34,23,65,23,5,99,18,15,14]
```

```python
import random
```

```python
for i in city_list:
  c1 = random.choice(city_list)
  c2 = random.choice(city_list)
  e1 = random.choice(edge_list)
  if c1 != c2:
    G4.add_edge(c1,c2, weight = e1)
G4.edges()
```

```python
nx.draw_networkx(G4)
plt.show()
```
